<h1><center>Analysis of Math and Reading Outcomes vs. School Resources</center></h1>

```python
import pandas as pd

schools_data = "Resources/schools_complete.csv"
students_data = "Resources/students_complete.csv"

# Read data files into DataFrames.
schools_df = pd.read_csv(schools_data)
students_df = pd.read_csv(students_data)
```

```python
# Print start of schools file
schools_df.head()
```

```python
# Print start of students file
students_df.head()
```

```python
# Merge the datasets with a left join.  
full_df = pd.merge(students_df, schools_df, how="left", on="school_name")
full_df.head()
```

```python
# Append three new columns to contain math and reading score status for later calculation.
full_df["passed_math"] = full_df["math_score"].apply(lambda m: m >= 70)
full_df["passed_reading"] = full_df["reading_score"].apply(lambda r: r >= 70)
full_df["passed_overall"] = full_df.apply(lambda x: x["math_score"] >= 70 and x["reading_score"] >= 70, axis = 1)
full_df.head()
```

```python
# Check key columns for typos and missing data. Remove any duplicates.
print(full_df["gender"].value_counts())
print(full_df["grade"].unique())
print(full_df["school_name"].unique())
print(full_df["reading_score"].unique())
print(full_df["math_score"].unique())
pd.DataFrame.drop_duplicates(full_df)
```

```python
# Print statistical overview of DataFrame.
full_df.describe()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Compute total schools, students, and budget and compute average math and reading scores for district.
total_schools = len(full_df["school_name"].unique())
total_students = len(full_df["student_name"])
total_budget = full_df["budget"].unique().sum()
avg_math = full_df["math_score"].mean()
avg_reading = full_df["reading_score"].mean()

# Compute percentage of students in district passing math and reading.
percent_passing_math = len(full_df.loc[full_df["math_score"] >= 70])/total_students*100
percent_passing_reading = len(full_df.loc[full_df["reading_score"] >= 70])/total_students*100
percent_passing_overall = len(full_df.loc[(full_df["math_score"] >=70) & (full_df["reading_score"] >= 70)])/total_students*100
```

```python
# Store district stats to DataFrame.
district_dict = [{"Total Schools": total_schools, "Total Students": total_students, "Total Budget": total_budget, "Average Math Score": avg_math, "Average Reading Score": avg_reading, "% Passing Math": percent_passing_math, "% Passing Reading": percent_passing_reading, "% Passing Overall": percent_passing_overall}]
district_df = pd.DataFrame(district_dict, columns = ["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Passing Overall"])
district_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Group full_df by school_name. Compute stats and aggregate to groupby object.
group_by_school = full_df.groupby("school_name").agg({"type": "first", "size": "first", "budget": "first", "math_score": "mean", "reading_score": "mean", "passed_math": "sum", "passed_reading": "sum", "passed_overall": "sum"})
group_by_school["budget_per_student"] = group_by_school["budget"]/group_by_school["size"]
group_by_school["%_passed_math"] = group_by_school["passed_math"]/group_by_school["size"]*100
group_by_school["%_passed_reading"] = group_by_school["passed_reading"]/group_by_school["size"]*100
group_by_school["%_passed_overall"] = group_by_school["passed_overall"]/group_by_school["size"]*100

# Convert groupby object to DataFrame and display.
group_by_school_df = pd.DataFrame(group_by_school)
group_by_school_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Store math scores by grade to series.
math_9th_gb = full_df.loc[full_df["grade"] == "9th"].groupby("school_name")
math_10th_gb = full_df.loc[full_df["grade"] == "10th"].groupby("school_name")
math_11th_gb = full_df.loc[full_df["grade"] == "11th"].groupby("school_name")
math_12th_gb = full_df.loc[full_df["grade"] == "12th"].groupby("school_name")

# Convert groupby objects to DataFrame objects
math_9th_df = math_9th_gb["math_score"].mean().to_frame().reset_index()
math_10th_df = math_10th_gb["math_score"].mean().to_frame().reset_index()
math_11th_df = math_11th_gb["math_score"].mean().to_frame().reset_index()
math_12th_df = math_12th_gb["math_score"].mean().to_frame().reset_index()

# Rename columns.
math_9th_df.rename(columns = {"math_score": "9th"}, inplace = True)
math_10th_df.rename(columns = {"math_score": "10th"}, inplace = True)
math_11th_df.rename(columns = {"math_score": "11th"}, inplace = True)
math_12th_df.rename(columns = {"math_score": "12th"}, inplace = True)

# Delect extra school_name columns.
del math_10th_df["school_name"]
del math_11th_df["school_name"]
del math_12th_df["school_name"]


# Concatenate DataFrame objects into a single DataFrame
math_by_grade_df = pd.concat([math_9th_df, math_10th_df, math_11th_df, math_12th_df], axis = 1)

math_by_grade_df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# Store reading scores by grade to series.
reading_9th_gb = full_df.loc[full_df["grade"] == "9th"].groupby("school_name")
reading_10th_gb = full_df.loc[full_df["grade"] == "10th"].groupby("school_name")
reading_11th_gb = full_df.loc[full_df["grade"] == "11th"].groupby("school_name")
reading_12th_gb = full_df.loc[full_df["grade"] == "12th"].groupby("school_name")

# Convert groupby objects to DataFrame objects.
reading_9th_df = reading_9th_gb["reading_score"].mean().to_frame().reset_index()
reading_10th_df = reading_10th_gb["reading_score"].mean().to_frame().reset_index()
reading_11th_df = reading_11th_gb["reading_score"].mean().to_frame().reset_index()
reading_12th_df = reading_12th_gb["reading_score"].mean().to_frame().reset_index()

# Rename columns.
reading_9th_df.rename(columns = {"reading_score": "9th"}, inplace = True)
reading_10th_df.rename(columns = {"reading_score": "10th"}, inplace = True)
reading_11th_df.rename(columns = {"reading_score": "11th"}, inplace = True)
reading_12th_df.rename(columns = {"reading_score": "12th"}, inplace = True)

# Delect extra school_name columns.
del reading_10th_df["school_name"]
del reading_11th_df["school_name"]
del reading_12th_df["school_name"]

# Concatenate DataFrame objects into a single DataFrame
reading_by_grade_df = pd.concat([reading_9th_df, reading_10th_df, reading_11th_df, reading_12th_df], axis = 1)
reading_by_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Group school data into 4 classes by budget_per_student and compare math and reading scores across classes.
bins = [578, 591.5, 628, 641.5, 655]
labels = ["Low", "Medium Low", "Medium High", "High"]
group_by_school_df["budget_per_student_class"] = pd.cut(group_by_school_df["budget_per_student"], bins, labels = labels, include_lowest = True)
group_by_budget_gb = group_by_school_df.groupby("budget_per_student_class").agg({"math_score": "mean", "reading_score": "mean", "%_passed_math": "mean", "%_passed_reading": "mean", "%_passed_overall": "mean"})
group_by_budget_gb.head()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Group school data into 4 classes by school size and compare math and reading scores across size classes.
bins = [427, 1858, 2949, 4635, 4976]
labels = ["Small", "Medium Small", "Medium Large", "Large"]
group_by_school_df["size_class"] = pd.cut(group_by_school_df["size"], bins, labels = labels, include_lowest = True)
group_by_size_gb = group_by_school_df.groupby("size_class").agg({"math_score": "mean", "reading_score": "mean", "%_passed_math": "mean", "%_passed_reading": "mean", "%_passed_overall": "mean"})
group_by_size_gb.head()
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
# Group school data into 2 classes by school type. Compare math and reading scores for district versus charter schools.
group_by_type_gb = group_by_school_df.groupby("type").agg({"math_score": "mean", "reading_score": "mean", "%_passed_math": "mean", "%_passed_reading": "mean", "%_passed_overall": "mean"})
group_by_type_gb.head()
```

<h3><center>Results<h3\><center\>

#### 1. There is a large disparity in budget per student between schools, ranging from 578 to 655 dollars. Schools with lower percentages of students passing tend to have larger budgets. That's counter-intuitive, but perhaps the budgets were increased in an attempt to improve student outcomes at under-performing schools.

#### 2. Comparing math and reading scores versus budget and school type it is clear that both math and reading are suffering in district schools with larger student bodies. However, there is a much larger difference in math scores than in reading scores. Perhaps more emphasis should be put into math at the under-performing schools.
